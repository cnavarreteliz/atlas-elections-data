```python
from epitools import get_data
import pandas as pd
```

```python
output = []
for election, year in [
    ("councilor", 2012),
    ("councilor", 2016),
    ("councilor", 2021)
    # ("first_round", 2013),
    # ("first_round", 2017),
    # ("first_round", 2021),
]:
    temp = get_data(
        country="Chile",
        year=year,
        election=election,
        aggregation="commune"
    )
    temp = temp.drop(columns=["epi", "epi_between"])
    output.append(temp)

df = pd.concat(output, ignore_index=True)
df["commune"] = df["commune"].replace({
    "CABO DE HORNOS(EX-NAVARINO)": "CABO DE HORNOS",
    "CABO DE HORNOS Y ANTARTICA": "CABO DE HORNOS",
    "AYSEN": "AISEN",
    "TREHUACO": "TREGUACO",
    "LLAY-LLAY": "LLAILLAY",
    "TIL TIL": "TILTIL",
    "OHIGGINS": "O´HIGGINS",
    "O'HIGGINS": "O´HIGGINS",
    "PAIHUANO": "PAIGUANO",
    "MARCHIGUE": "MARCHIHUE",
    "COYHAIQUE": "COIHAIQUE",
    "OLLAGUE": "OLLAGÜE"
})

df = df.rename(columns={"commune": "Comuna"})
# {'AISEN', 'COIHAIQUE', 'MARCHIHUE', 'OLLAGÜE', 'O´HIGGINS', 'TREGUACO'}
df.head()
```

```python
dd = df.pivot(index="Comuna", columns="year", values="epi_within")
dd = dd.reset_index()
# dd.columns = [
#     "Comuna",
#     "Presidencial 2013 (EP)",
#     "Presidencial 2017 (EP)",
#     "Presidencial 2021 (EP)",
#     "Presidencial 2013 (Between-EP)",
#     "Presidencial 2017 (Between-EP)",
#     "Presidencial 2021 (Between-EP)",
#     "Presidencial 2013 (Within-EP)",
#     "Presidencial 2017 (Within-EP)",
#     "Presidencial 2021 (Within-EP)",
# ]
dd.head()
```

```python
data = pd.read_csv(
    "https://docs.google.com/spreadsheets/d/e/2PACX-1vSSncoe7MOFOpcv51GlTRJxNO-Nbd7ORwvTVfFah3YMM7Td2i0dFwHLea5UIuDbiA/pub?output=csv"
)
for a, b in [("Á", "A"), ("É", "E"), ("Í", "I"), ("Ó", "O"), ("Ú", "U")]:
    data["Comuna"] = data["Comuna"].str.replace(a, b)

data.head()
```

```python
pd.merge(
    data[["ID_antiguo", "ID_nuevo", "Comuna", "provincia", "Region"]],
    dd,
    on="Comuna",
    how="outer"
).to_excel("Datos_EP_Chile.xlsx", index=False)
```
