```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd
import statsmodels.api as sm
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
# Read full data
df = pd.read_excel("data/France/2022_first_round.xlsx")
df.head()
```

```python
df = df.rename(columns={
    "Voix": "Nathalie ARTHAUD",
    "Unnamed: 32": "Fabien ROUSSEL",
    "Unnamed: 39": "Emmanuel MACRON",
    "Unnamed: 46": "Jean LASSALLE",
    "Unnamed: 53": "Marine LE PEN",
    "Unnamed: 60": "Éric ZEMMOUR",
    "Unnamed: 67": "Jean-Luc MÉLENCHON",
    "Unnamed: 74": "Anne HIDALGO",
    "Unnamed: 81": "Yannick JADOT",
    "Unnamed: 88": "Valérie PÉCRESSE",
    "Unnamed: 95": "Philippe POUTOU",
    "Unnamed: 102": "Nicolas DUPONT-AIGNAN"
})
```

```python
df_filtered = df[[
    "Code du département", 
    "Libellé de la commune", 
    "Code du b.vote", 
    "Nathalie ARTHAUD", 
    "Fabien ROUSSEL", 
    "Emmanuel MACRON",
    "Jean LASSALLE", 
    "Marine LE PEN", 
    "Éric ZEMMOUR", 
    "Jean-Luc MÉLENCHON", 
    "Anne HIDALGO",
    "Yannick JADOT",
    "Valérie PÉCRESSE",
    "Philippe POUTOU",
    "Nicolas DUPONT-AIGNAN"
]]

df_filtered = df_filtered.melt(
    id_vars=["Code du département", "Libellé de la commune", "Code du b.vote"], 
    var_name="Candidate", 
    value_name="Votes"
)

df_filtered["polling_id"] = df_filtered["Libellé de la commune"] + "-" + df_filtered["Code du b.vote"]

df_filtered
```

```python
#[df_filtered["Code du département"] == "75"]
df_1st_round = df_filtered\
    .groupby(["Code du département", "Libellé de la commune", "ID", "Candidate"]).agg({"Votes": "sum"})\
    .sort_values(["Code du département", "Libellé de la commune", "ID", "Votes"], ascending=[False, True, True, False])
df_1st_round["Weight"] = df_1st_round.groupby(level=[0, 1, 2]).apply(lambda x: x/x.sum())
df_1st_round["Rank"] = df_1st_round.groupby(level=[0, 1, 2]).cumcount() + 1
df_1st_round = df_1st_round.reset_index()
df_1st_round.head()
```

```python
df_1st_round
```

```python
dd = df_1st_round.groupby(["Code du département", "ID", "Candidate"]).agg({"Rank": "sum"})


output = []
for i, tmp in dd.groupby(level=[0, 1]):
    department_id, voter = i
    s = tmp.sort_values("Rank").reset_index()
    
    output.append({
        "department_id": department_id,
        "voter": voter,
        "rank": ">".join(s["Candidate"])
    })
    
data = pd.DataFrame(output)
data.shape
```

```python
data.to_csv("france_1st_round_pseudo_voter_data.csv", index=False)
```

```python
df_pop = df.groupby("Code du département").agg({"Inscrits": "sum", "Abstentions": "sum"}).reset_index()
df_pop["Abstention Rate"] = df_pop["Abstentions"] / df_pop["Inscrits"]
df_pop
```

```python

df_model = pd.merge(dd, df_gini2, left_on="Code", right_on="CODGEO")
df_model = pd.merge(df_model, df_pop, on="Code du département")
df_model["Taux de pauvreté (en\xa0%)"] = df_model["Taux de pauvreté (en\xa0%)"] / 100
df_model["Inscrits Log"] = np.log10(df_model["Inscrits"])
df_model.corr()
```

```python
Y = df_model["Weight"]
X = df_model[["Taux de pauvreté (en\xa0%)", "GI16", "Inscrits Log", "Abstention Rate"]]
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit(cov_type="hc2")

print(results.summary())
```
