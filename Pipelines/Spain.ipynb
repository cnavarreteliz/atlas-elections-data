```python
country = "Spain"
month = 11
year = 2019
election_type = "general"
```

```python
output = []
path = f"data/{country}/{year}{month}.zip"
for i in ZipFile(path).namelist():
    print(i)
    if i[-3:] == "DAT":
        tmp = pd.read_csv(ZipFile(path).open("09021911.DAT"), on_bad_lines="skip", header=None, encoding="latin-1")
        output.append(tmp)
```

```python
df = pd.concat(output, ignore_index=True)
```

```python
columns = [(1, 2, "election_type"), (3, 6, "year"), (7, 8, "month"), (9, 9, "n_round"),
(10, 11, "state_id"), (12, 13, "province_id"), (14, 16, "commune_id"), (17, 18, "district_commune_id"),
(19, 22, "section_id"), (23, 23, "polling_code"), (24, 29, "candidate_id"), (30, 36, "value")]

df = pd.read_csv(ZipFile(path).open("10021911.DAT"), header=None)

for column in columns:
    df[column[2]] = df[0].str[column[0] - 1:column[1]]
    
df["polling_id"] = df["state_id"] + df["province_id"] + df["commune_id"] + df["district_commune_id"] + df["section_id"] + df["polling_code"]
df[["candidate_id", "value"]] = df[["candidate_id", "value"]].astype(int)
df.head()
```

```python
columns = [(1, 2, "election_type"), (3, 6, "year"), (7, 8, "month"), (9, 14, "candidate_id"),
(15, 64, "candidate"), (65, 214, "candidate_description"), (215, 220, "candidate_code_province"), 
(221, 226, "candidate_code_state"), (227, 232, "candidate_code_country")]

df_candidates = pd.read_csv(ZipFile(path).open("03021911.DAT"), encoding="latin-1", header=None)

for column in columns:
    df_candidates[column[2]] = df_candidates[0].str[column[0] - 1:column[1]]

df_candidates[["candidate_id"]] = df_candidates[["candidate_id"]].astype(int)
df_candidates["candidate"] = df_candidates["candidate"].str.strip()
df_candidates["candidate_description"] = df_candidates["candidate_description"].str.strip()

df_candidates.head()
```

```python
df_candidates[["candidate_id", "candidate"]].tail(50)
```

```python
df = pd.merge(df, df_candidates[["candidate_id", "candidate"]], on="candidate_id")
df["state_id"] = df["state_id"].astype(int)
df.head()
```

```python
df.state_id.unique()
```

```python
df_filtered = df[["candidate", "value", "polling_id"]].copy()
df_filtered["rank"] = df_filtered.groupby(["polling_id"])["value"].rank("min", ascending=False).astype(int)
```

```python
df.candidate.unique()
```

```python
df_filtered = df_filtered.groupby(["polling_id", "candidate"]).agg({"value": "sum"})
df_filtered["rate"] = df_filtered.groupby(level=[0], group_keys=False).apply(lambda x: x/x.sum())
df_filtered = df_filtered.reset_index()
df_filtered = df_filtered[["polling_id", "candidate", "rate", "value"]]
```

```python
df_nuts = pd.read_csv("data/Spain/nuts_2_equivalences.csv")
df_nuts["nuts_1"] = df_nuts["nuts_2"].str[0:3]
df_nuts.head()
```

```python
df_filtered.to_csv(f"data_output/Spain/{year}_{election_type}.csv.gz", compression="gzip", index=False)
```

```python
df_location = df[["polling_id", "state_id", "province_id", "commune_id", "district_commune_id"]]
df_location = pd.merge(df_location, df_nuts, how="left", on="state_id")
df_location.to_csv(f"data_output/Spain/{year}_{election_type}_location.csv.gz", compression="gzip", index=False)
```
