```python
import pandas as pd
# https://computos2018.ine.mx/#/descargaBase
```

```python
year = 2018
election = "first_round"
```

```python
df = pd.read_csv(f"../data/Mexico/{year}_first_round.csv", encoding="latin-1", delimiter="|", header=5)
df = df.dropna(axis=1, how="all")
df
```

```python
len(df["CLAVE_CASILLA"].unique())
```

```python
df_candidates = pd.read_csv(f"../data/Mexico/{year}_first_round_candidates.csv", encoding="latin-1", delimiter="|", header=1)
party_columns = df_candidates["PARTIDO_CI"].unique()
candidacies = df_candidates[["PARTIDO_CI", "CANDIDATURA_PROPIETARIA"]].set_index("PARTIDO_CI").to_dict()["CANDIDATURA_PROPIETARIA"]
df_candidates
```

```python
set(list(df)) - set(party_columns)
```

```python
df_tmp = df[
    (set(["CLAVE_CASILLA", "CAND_IND_01", "CAND_IND_02", "CNR", "ENCUENTRO SOCIAL", "MOVIMIENTO CIUDADANO", "VN"]) | set(party_columns)) 
    - set(['MC', 'PES', 'CI_39987', 'CI_76991'])]\
    .melt(id_vars=["CLAVE_CASILLA"], value_name="value", var_name="candidate")

df_tmp["value"] = df_tmp["value"].replace("-", 0).astype(int)
df_tmp["candidate"] = df_tmp["candidate"].replace({
    "ENCUENTRO SOCIAL": "PES",
    "MOVIMIENTO CIUDADANO": "MC",
    "CNR": "OTHER", # NOT REGISTERED CANDIDACIES
    "VN": "SPOILT",
    "CAND_IND_01": "CI_39987",
    "CAND_IND_02": "CI_76991"
}).replace(candidacies)
df_tmp = df_tmp.rename(columns={"CLAVE_CASILLA": "polling_id"})
df_tmp = df_tmp.groupby(["polling_id", "candidate"]).agg({"value": "sum"}).reset_index()

## Verifies whether the total number of votes coincides with the reported by the INE
assert df_tmp.value.sum() == 56611027
```

```python
tt = df_tmp.groupby(["polling_id", "candidate"]).agg({"value": "sum"})
tt["rate"] = tt.groupby(level=[0], group_keys=False).apply(lambda x: x/x.sum())
tt = tt.reset_index()

tt["rank"] = tt.groupby("polling_id")["value"].rank("min", ascending=False).astype(int)

tt
```

```python
tt.to_csv(f"../data_output/Mexico/{year}_{election}.csv.gz", compression="gzip", index=False)
```

```python
df_location = df[["CLAVE_CASILLA", "ID_ESTADO", "NOMBRE_ESTADO", "ID_DISTRITO", "NOMBRE_DISTRITO"]]\
    .rename(columns={
        "CLAVE_CASILLA": "polling_id",
        "ID_ESTADO": "state_id",
        "NOMBRE_ESTADO": "state",
        "ID_DISTRITO": "district_id",
        "NOMBRE_DISTRITO": "district",
        "SECCION": "section"
    }).drop_duplicates()

df_location.to_csv(f"../data_output/Mexico/{year}_{election}_location.csv.gz", compression="gzip", index=False)
```
