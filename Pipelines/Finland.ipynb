```python
import pandas as pd
import math
```

```python
storage_options = {"User-Agent": "Mozilla/5.0"}

d95 = "http://tilastokeskus.fi/tk/he/vaalit/vaalit95/risvaa.xls"
d99 = "http://tilastokeskus.fi/tk/he/vaalit/vaalit99/ehdokas.xls"
d03 = "http://tilastokeskus.fi/tk/he/vaalit/vaalit2003/vaalit2003_vaalitilastot_ehdokkaatverluku.xls"
d07 = "http://www.stat.fi/til/evaa/2007/evaa_2007_2007-03-28_tau_015.xls"
df = pd.read_excel(d03, header=6, storage_options=storage_options)
df = df.iloc[:, : 3]
df.head()
```

```python
df.columns = ["item", "value", "comparison"]
df = df[1:]
df = df.dropna(subset=["item"])
df["value"] = df["value"].apply(lambda x: str(x).strip()).replace(" ", pd.np.nan).replace("", pd.np.nan).astype(float)
df.head()
```

```python
district_id = None
party_id = None

parties = list(df["item"].value_counts()[df["item"].value_counts() > 2].index)
output = []
for i, item in df.iterrows():
  if item["item"] in parties and math.isnan(item["value"]):
    party_id = item["item"]
  elif item["item"] not in parties and math.isnan(item["value"]):
    district_id = item["item"]

  if not math.isnan(item["value"]): 
    output.append({
      "district_id": district_id,
      "party_id": party_id,
      "candidate": item["item"].replace("...", ""),
      "value": int(item["value"]),
    })

data = pd.DataFrame(output)
data.to_csv(f"../data_output/Finland/{2003}_camera_district.csv.gz", index=False, compression="gzip")

data.head()
```
