```python
import geopandas as gpd
import numpy as np
import pandas as pd
```

```python
year = 2016
election = "senate"
```

```python
df = pd.read_csv(f"../data/United States/{year}_{election}_precinct_general.csv", encoding="ISO-8859-1")
# df[["county_fips", "jurisdiction_fips"]] = df[["county_fips", "jurisdiction_fips"]].fillna(-1).astype(int)
df = df.rename(columns={"state_postal": "state_po", "jurisdiction": "jurisdiction_name"})
df.head()
```

```python
# list(df)
```

```python
def create_polling_id(data, columns=[]):
    output = data[columns[0]].copy()

    for column in columns[1:]:
        output += "-" + data[column].astype(str)
    return output
```

```python
df["polling_id"] = create_polling_id(df, ["state_po", "county_fips", "jurisdiction_name", "district", "precinct", "mode"])
```

```python
df.head()
```

```python
list(df)
```

```python
if election == "senate":
    df = df.rename(columns={"candidate": "candidate_name", "party_simplified": "candidate", "party": "candidate"})
```

```python
df["rank"] = df.groupby(["polling_id"])["votes"].rank("min", ascending=False).astype(int)
```

```python
# list(df)
```

```python
tt = df.groupby(["polling_id", "candidate"]).agg({"votes": "sum"})
tt["rate"] = tt.groupby(level=[0], group_keys=False).apply(lambda x: x/x.sum()).fillna(0)
tt = tt.reset_index()
tt = tt[["polling_id", "candidate", "rate"]]
```

```python
df = pd.merge(df, tt, on=["polling_id", "candidate"])
```

```python
df.candidate.unique()
```

```python
df.loc[(df["state"] == "MINNESOTA") & (df["party_detailed"] == "DEMOCRATIC FARMER LABOR"), "candidate"] = "DEMOCRAT"
```

```python
# df[df["state"] == "ARKANSAS"][["polling_id", "candidate", "votes", "rank", "rate"]]
```

```python
df_filtered = df[["polling_id", "candidate", "votes", "rank", "rate"]].copy()
df_filtered = df_filtered.rename(columns={"votes": "value"})
df_filtered["candidate"] = df_filtered["candidate"].apply(lambda x: x.strip())

df_filtered.to_csv(f"../data_output/United States/{year}_{election}.csv.gz", compression="gzip", index=False)
```
