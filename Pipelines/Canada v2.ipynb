```python
import zipfile
import pandas as pd
from io import BytesIO
import requests

links = {
    2021: "https://www.elections.ca/res/rep/off/ovr2021app/53/data_donnees/pollresults_resultatsbureauCanada.zip"
}

result = requests.get(links[2021])

output = []
with zipfile.ZipFile(BytesIO(result.content)) as z:
    files = z.namelist()
    for f in files:
        frag = pd.read_csv(z.open(f))
        output.append(frag)

df = pd.concat(output, ignore_index=True)
df.head()
```
