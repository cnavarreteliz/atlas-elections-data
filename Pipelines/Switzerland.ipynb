```python
import pandas as pd

pd.options.display.max_columns = None
```

```python
year = 2015

df = pd.read_csv(f"../data/Switzerland/{year}.csv")
df["candidate"] = df["candidate_first_name"] + " " + df["candidate_family_name"]
df = df.rename(columns={
  "candidate_votes": "value",
  "list_name": "party_id",
  "entity_district": "district_id",
})
# df = df[["candidate", "value", "party_id", "district_id"]]
df.head()
```

```python
election_type = "camera"

df.to_csv(f"../data_output/Switzerland/{year}_{election_type}_district.csv.gz", index=False, compression="gzip")

df.head()
```
