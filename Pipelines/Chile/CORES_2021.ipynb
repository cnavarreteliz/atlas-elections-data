```python
import pandas as pd
```

```python
df = pd.read_csv("2021/cores/2021_11_ConsejerosRegionales_Votacion.zip", delimiter="|")
df["Candidato"] = df["Nombres"] + " " + df["Primer apellido"]+ " " + df["Segundo apellido"]
df.head()
```

```python
columns = {
  "Region": "region",
  "Circunscripcion provincial": "district",
  "Comuna": "commune",
  "Local": "place",
  "Mesa": "polling_place",
  "Pacto": "coalition",
  "Subpacto": "subcoalition",
  "Partido": "party",
  "Candidato": "candidate",
  "votos": "value"
}
```

```python
df = df.rename(columns=columns)
df = df[columns.values()]
```

```python
def create_polling_id(data, columns=[]):
  output = data[columns[0]].fillna("###").copy()

  for column in columns[1:]:
    output += "-" + data[column].fillna("###").astype(str)
  return output

cols = ["region", "district", "commune", "place", "polling_place"]
df["polling_id"] = create_polling_id(df, columns=cols)
```

```python
df[df.candidate.fillna("#").str.contains("GABRIEL MAURICIO TORRES")]
```

```python
data = df.dropna(subset=["coalition"]).copy()
data["value"] = data["value"].astype(str).apply(lambda x: x.strip()).apply(lambda x: 0 if x == "" else x).astype(float)
data["subcoalition"] = data["subcoalition"].fillna(data["coalition"])
tt = data.groupby(["polling_id", "coalition", "subcoalition", "party", "candidate"]).agg({"value": "sum"})
tt["rate"] = tt.groupby(level=[0], group_keys=False).apply(lambda x: x/x.sum()).fillna(0)
tt = tt.reset_index()
tt["flag_candidates"] = 1

tt.head()
```

```python
df_location = df[["polling_id", "region", "district", "commune", "place", "polling_place"]].drop_duplicates()
```
