```python
import pandas as pd
import requests

pd.options.display.max_columns = None
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36"}

coalitions = requests.get("https://resultadoshistorico.onpe.gob.pe/assets/json/agpcand/D44001.json", headers=headers).json()
coalitions = pd.DataFrame(coalitions["organizaciones"])
coalitions.head()
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36"}

data = requests.get("https://resultadoshistorico.onpe.gob.pe/assets/json/distritoelectoral/11.json", headers=headers).json()
districts = pd.DataFrame(data["distritoelectoral"])
districts.head()
```

```python
output = []
for party_id in coalitions["C_CODI_AGRUPOL"].unique():
    for district_id in districts["CDGO_DEP"].unique():
        data = requests.get(f"https://resultadoshistorico.onpe.gob.pe/assets/json/results/11/{district_id}/{party_id}.json", headers=headers).json()
        frag = pd.DataFrame(data["results"])
        frag["district_id"] = district_id
        frag["party_id"] = party_id
        output.append(frag)
```

```python
data = pd.concat(output, ignore_index=True)
data = data.rename(columns={"C_CODI_CANDIDATO": "candidate_id", "CANDIDATO": "candidate", "NUMLISTA": "position", "TOTAL_VOTOS": "value"})
data["value"] = data["value"].str.replace(",", "").astype(int)

election = "lower"
data.to_csv(f"../../data_output/Peru/{2021}_{election}.csv.gz", compression="gzip", index=False)
data.head()
```
