```python
df_voters = df_location.groupby("polling_id").agg({"voters": "sum"}).reset_index()
df_voters.head()
```

```python
df_part_fround = df.groupby("polling_id").agg({"value": "sum"}).reset_index()
df_part_runoff = df_runoff.groupby("polling_id").agg({"value": "sum"}).reset_index()

df_part_fround = pd.merge(df_voters, df_part_fround, on="polling_id")
df_part_fround["abstention_fround"] = df_part_fround["value"] / df_part_fround["voters"]

df_part_runoff = pd.merge(df_voters, df_part_runoff, on="polling_id")
df_part_runoff["abstention_runoff"] = df_part_runoff["value"] / df_part_runoff["voters"]
```

```python
df_tmp = df.groupby([location_level, "candidate"]).agg({"value": "sum"})
df_tmp["rate"] = df_tmp.groupby(level=[0]).apply(lambda x: x/x.sum())
df_tmp = df_tmp.reset_index()
df_tmp.head()
```

```python
df_hhi = df_tmp.copy()
df_hhi["hhi"] = df_hhi["rate"] ** 2
df_hhi["entropy"] = df_hhi["rate"] * np.log(df_hhi["rate"])
df_hhi = df_hhi.groupby(location_level).agg({"hhi": "sum", "entropy": "sum"})
df_hhi["entropy"] = -1 * df_hhi["entropy"]
df_hhi = df_hhi.reset_index()
df_hhi.head()
```

```python
location = "polling_id"

def get_rate(df, location=["polling_id"], level=[0]):
    df_tmp = df.groupby(location + ["candidate"]).agg({"value": "sum"}) # [~df["candidate"].isin(values)]
    df_tmp["rate"] = df_tmp.groupby(level=level).apply(lambda x: x/x.sum())
    df_tmp = df_tmp.reset_index()

    return df_tmp

df_a = get_rate(df, location=[location_level, "polling_id"], level=[0, 1])
df_a = pd.merge(df_a, df_dv.rename(columns={"value": "divisiveness"}), on=[location_level, "candidate"])

df_b = get_rate(df_runoff)
```
